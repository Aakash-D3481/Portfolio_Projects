```python
# Importing the Necessary Packages

import os
import glob
import re
import pandas as pd
import warnings 
from natsort import natsorted
warnings.filterwarnings("ignore")


# Provide the Directory Path for where the Excel Files are stored

path = r'C:\Users\Aakash D\Downloads\UK VS 2022\PROCESSED'
```

## For Checking All the Headers in the FORM 20

```python
flag = 0
Bad_files = []

folder_path = path

excel_files = glob.glob(os.path.join(folder_path, '*.xls*'))
file_count = 0

for excel_file in excel_files:
    file_count += 1
    file_name = os.path.basename(excel_file)
    print("Processing File ", file_name)

    df = pd.read_excel(excel_file)

    if "Constituency" in df.columns:
        con = None
    else:
        con = "Constituency Issue"
        flag += 1
        Bad_files.append(f"{file_name}, issue is {con}")
        continue

    if "P.S.No" in df.columns:
        ps = None
    else:
        ps = "P.S.No Issue"
        flag += 1
        Bad_files.append(f"{file_name}, issue is {ps}")
        continue

    if "Total Votes" in df.columns:
        tv = None
    else:
        tv = "Total Votes Issue"
        flag += 1
        Bad_files.append(f"{file_name}, issue is {tv}")
        continue

    if "Total of valid votes" in df.columns:
        tof = None
    else:
        tof = "Total valid votes issue"
        flag += 1
        Bad_files.append(f"{file_name}, issue is {tof}")
        continue
    
    if "No of tendered votes." in df.columns:
        ntv = None
    else:
        ntv = "Tendered vote Issue"
        flag += 1
        Bad_files.append(f"{file_name}, issue is {ntv}")
        continue
    
    if "SI.No." in df.columns:
        sin = None
    else:
        sin = "serial issue"
        flag += 1
        Bad_files.append(f"{file_name}, issue is {sin}")
        continue
    
    if "No.of rejected votes" in df.columns:
        rej = None
    else:
        rej = "Rejected Issue"
        flag += 1
        Bad_files.append(f"{file_name}, issue is {rej}")
        continue

Bad_files = natsorted(Bad_files)
print("Total Files Processed are ", file_count)
print("Total Flags Detected are ", flag)
for i in Bad_files:
    print(i)
```

## Check For Null Values

```python
bad_files = []


folder_path = path


excel_files = glob.glob(os.path.join(folder_path, '*.xls*'))


for excel_file in excel_files:
    file_name = os.path.basename(excel_file)
    print(f"Processing File: {file_name}")
    df = pd.read_excel(excel_file)

    null_found = sum(df.isna().sum().values)

    if null_found == 0:
        None
    else:
        print("Nulls FOUND")
        bad_files.append(file_name)
        df.fillna(0, inplace=True)

bad_files = natsorted(bad_files)

if len(bad_files) > 0:
    for i in bad_files:
        print(i)
        
else:
    print("\nNo Nulls Found, Good to Go!")
    
```

## Check for NEGATIVE Values in "NOTA" Column

```python
folder_path = path

excel_files = glob.glob(os.path.join(folder_path, '*.xls*'))
file_count = 0
bad_files = []
flag = 0

for excel_file in excel_files:
    try:
        file_count += 1
        file_name = os.path.basename(excel_file)
        print("Processing File ", file_name)

        df = pd.read_excel(excel_file)

        for index, row in df.iterrows():
            if row['NOTA'] < 0:
                print("Negatives Detected")
                flag = 1

        if flag == 1:
            bad_files.append(file_name)
            flag = 0
    except Exception as e:
        print(f"Exception Occurred for {file_name}, exception is:  {e}")
        bad_files.append(f'{file_name} with exception {e}')

print("The Number of Bad Files are ", len(bad_files))
bad_files = natsorted(bad_files)

for i in bad_files:
    print(i)
```

## To Check For any String Values in the Votes Data

```python
folder_path = path

bad_files = []

excel_files = glob.glob(os.path.join(folder_path, '*.xls*'))

for excel_file in excel_files:
    try:
        file_name = os.path.basename(excel_file)
        print("Processing ", file_name)
        df = pd.read_excel(excel_file)
        flag = 0

        skip_cols = ['SI.No.', 'P.S.No', 'Constituency', 'Total of valid votes', 'No.of rejected votes', 'Total Votes', 'No of tendered votes.', 'NOTA']

        cols = [col for col in df.columns if col not in skip_cols]

        tally_df = df.loc[:,cols]

        add_cols_df = df[['NOTA', 'Total Votes']]

        add_cols_df

        check_df = pd.concat([tally_df, add_cols_df], axis=1)
        check_df[[col for col in check_df.columns]] = check_df[[col for col in check_df.columns]].astype(int)

        for index, row in check_df.iterrows():
            party_sum_plus_nota = sum(row[:-1])
            total_votes = row.iloc[-1]

            if party_sum_plus_nota != total_votes:
                flag += 1

        if flag > 0:
            print("MisMatch Detected")
            bad_files.append(file_name)
    except Exception as e:
        print("Error Occurred", e)
        bad_files.append(f"{file_name}, has exception as {e}")

print(flag)

bad_files = natsorted(bad_files)
for i in bad_files:
    print(i)

print(f'\nTotal Files with issues are: {len(bad_files)}')
```

## Check For any Non Numeric Characters in Tendered and Rejected Votes Columns

```python
folder_path = path

bad_files = []

excel_files = glob.glob(os.path.join(folder_path, '*.xls*'))
flag = 0
for excel_file in excel_files:
    try:
        file_name = os.path.basename(excel_file)
        print("Processing file: ", file_name)
        df = pd.read_excel(excel_file)
        for _, row in df.iterrows():
            try:
                pd.to_numeric(row["No of tendered votes."])
                pd.to_numeric(row["No.of rejected votes"])
            except ValueError:
                print("Non-numeric character detected")
                flag += 1
        if flag == 0:
            print("Check Successful")
        else:
            bad_files.append(file_name)
            flag = 0
    except Exception as e:
        print("Exception occurred ", e)

print("________________________________________________________________________________________")
bad_files = natsorted(bad_files)
for i in bad_files:
    print(i)
```

```python
original_directory_path = path
ECI_Detailed_Report_Path = r"C:\Users\Aakash D\Downloads\UK VS 2022\Uk_2022_Detailed Results.xlsx"
# Details for Column names in ECI Report

#incase of LK
state_col = 'STATE/UT NAME'
state_name = 'Assam'
#incase of LK and VS
ac_num = 'AC NO.'
constituency = 'AC NAME'
candidate_names = 'CANDIDATE NAME'
party_names = 'PARTY'
total_votes = 'GENERAL'
```

## Comparison of The Votes To ensure Integrity

```python
original_lst = []

edited_folder_path = os.path.join(os.getcwd(), "Edited_Files")
os.makedirs(edited_folder_path, exist_ok=True)

folder_path = original_directory_path

excel_files = glob.glob(os.path.join(folder_path, '*.xls*'))
file_count = 0

for excel_file in excel_files:
    file_count += 1
    file_name = os.path.basename(excel_file)
    print("Processing File: ", file_name)
    num = re.search(r'\d+', file_name).group()
    num = int(num)
    df = pd.read_excel(excel_file)
    df['File_Num'] = num

    edited_file_path = os.path.join(edited_folder_path, f'{file_name}_edited.xlsx')
    df.to_excel(edited_file_path, index=False)

print("\nThe Files were resaved in Edited_Files Folder")
print("\nTotal Files Processed & Converted: ", file_count)

edited_folder_path = os.path.join(os.getcwd(), 'Edited_Files')

# Specify the directory containing your Excel files
excel_files_directory = edited_folder_path

# Specify the pattern for Excel files (e.g., all .xlsx files)
excel_files_pattern = '*.xls*'

# Get a list of file paths matching the pattern
excel_files = glob.glob(os.path.join(excel_files_directory, excel_files_pattern))

# Initialize an empty DataFrame to store the appended data
appended_df = pd.DataFrame()

# Loop through each Excel file
for excel_file in excel_files:
    try:
        # Read the Excel file into a DataFrame
        df = pd.read_excel(excel_file)

        # Convert from wide to long form using melt
        long_df = pd.melt(df, id_vars=['P.S.No', 'Constituency','File_Num'], var_name='Variable', value_name='Value')

        # Append the long form DataFrame to the overall DataFrame
        appended_df = pd.concat([appended_df, long_df], ignore_index = True)

        print(f'DataFrame from {excel_file} converted and appended.')
    except FileNotFoundError:
        print(f'Error: File {excel_file} not found.')
    except Exception as e:
        print(f'Error in file {excel_file}: {e}')

# Save the appended DataFrame to a single Excel file
output_file_path = os.path.join(excel_files_directory, 'Complete_Long_Form.csv')
appended_df.to_csv(output_file_path, index=False)
print(f'Appended DataFrame saved to {output_file_path}')

edited_folder_path = os.path.join(os.getcwd(), "Edited_Files")

long_form_path = os.path.join(edited_folder_path, "Complete_Long_Form.csv")

df = pd.read_csv(long_form_path, low_memory=False)

df = df[~df['Variable'].isin(["NOTA", "No of tendered votes.", "No.of rejected votes", "Total of valid votes", "Total Votes", "SI.No."])]

df = df.drop_duplicates()

df['Value'] = df['Value'].astype(int)

df["Total_by_Constituency"] = df.groupby('Constituency')['Value'].transform('sum')
df["Total_by_Cons_and_parties"] = df.groupby(['Constituency', 'Variable'])['Value'].transform('sum')

filtered_df = df[['File_Num','Constituency', 'Variable', 'Total_by_Constituency', 'Total_by_Cons_and_parties']]
filtered_df = filtered_df.drop_duplicates().reset_index(drop = True)

print(f"Total Constituencies Detected are: {filtered_df['Constituency'].value_counts().count()}")

# For Reference File
ref = pd.read_excel(ECI_Detailed_Report_Path)

# Uncomment the below code incase of LOK SABHA
# ref = ref[ref[state_col].isin([state_name])]

folder_path = original_directory_path

original_lst = []

excel_files = glob.glob(os.path.join(folder_path, '*.xls*'))

for excel_file in excel_files:
    file_name = os.path.basename(excel_file)
    num = re.search(r'\d+', file_name).group()
    num = int(num)
    original_lst.append(num)
    original_lst = sorted(original_lst)
    large = max(original_lst)
    check_lst = list(range(1, large+1))

    missing_files = [file for file in check_lst if file not in original_lst]
    missing_files = sorted(missing_files)
print(missing_files)

exclusion_lst = missing_files
ref = ref[~ref[ac_num].isin(exclusion_lst)]

ref = ref[[ac_num, constituency, party_names, total_votes]]
ref.dropna(inplace=True)
ref[[ac_num, total_votes]] = ref[[ac_num, total_votes]].astype(int)

# Dealing with NOTA if Present in Party NAME
exclude = ['NOTA']
ref = ref[~ref[party_names].isin(exclude)]

ref['Ref_Con_Total'] = ref.groupby(constituency)[total_votes].transform('sum')

# Sorting & Concatenation

filtered_df = filtered_df.sort_values(by=['File_Num','Total_by_Cons_and_parties']).reset_index(drop=True)
ref = ref.sort_values(by=[ac_num, total_votes]).reset_index(drop=True)

party_concat = pd.concat([ref, filtered_df], axis=1)
party_concat.fillna('0', inplace=True)
party_concat[[ac_num, total_votes,'Ref_Con_Total','File_Num','Total_by_Constituency','Total_by_Cons_and_parties']] = party_concat[[ac_num, total_votes,'Ref_Con_Total','File_Num','Total_by_Constituency','Total_by_Cons_and_parties']].astype(int)

#Conversion To Excel
party_concat.to_excel('Party_concat.xlsx', index=False)

party_concat.head(10)
```
